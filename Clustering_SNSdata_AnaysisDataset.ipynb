```python
import pandas as pd
```

```python
snsdata=pd.read_csv("snsdata.csv")
```

```python
snsdata.head()
```

```python
snsdata.isnull().sum()
```

```python
snsdata.shape,snsdata['gradyear'].value_counts()
```

```python
snsdata.describe()
```

```python
X=snsdata[["gradyear"]]
```

```python
from sklearn.cluster import KMeans
```

```python
clf=KMeans(n_clusters=3)
clf.fit(X)
clf.cluster_centers_
```

```python
intrests=snsdata[['football','music']].sample(10000)
intrests
```

```python
clf=KMeans(n_clusters=2)
clf.fit(intrests)
clf.cluster_centers_
```

```python
intrests['cluster']=clf.predict(intrests)
```

```python
intrests
```

```python
import matplotlib.pyplot as plt
plt.scatter(x='football',y='music',data=intrests[intrests.cluster==0],marker='o',color='red',label='cluster 0')
plt.scatter(x='football',y='music',data=intrests[intrests.cluster==1],marker='o',color='blue',label='cluster 1')
plt.legend()
plt.xlabel("Football")
plt.ylabel("Music")
plt.grid()
```
